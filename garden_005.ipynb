```python
#@title Garden

garden_target = target_input
# right_target = garden_target//2
# left_target = garden_target//2
# # left_count = 9
# # right_count = 10

def cc(side):
  #count_converter
  ratio = side/(garden_target/2)
  if ratio < 0.2:
    state = 'dead'
  elif ratio < 0.6:
    state = 'dormant'
  elif ratio < 1:
    state = 'healthy'
  elif ratio >=1:
    state = 'thriving'
  return state

description = {'dead': 'There is nothing on this side but \n a few rocks. Try reading more articles \n from this side of the political \n spectrum to plant some seedlings.',
               'dormant': 'This side of the lawn is very dry \n but you have started to water it \n and plant some seedlings. Read more \n articles from this side of the political \n spectrum to see your plants flower.', 
               'healthy': 'This side of the lawn is still a little \n pale but your plants have begun to \n flower. Keep reading more articles \n from this side of the political spectrum \n to see your plants fully mature.',
               'thriving':'This side of the lawn is thriving. \n You are reading as much as you aimed to \n from this side of the political spectrum.'}

def trianglify(x,y,s):
    return [[x-0.5*s,y-0.5*s], [x+0.5*s,y-0.5*s], [x, y+0.5*s]]

def make_flower(x,y,c):
    e11 = patches.Ellipse(xy=(x,y), width=5, height=2, fc=c, lw=None, zorder=10)
    e12 = patches.Ellipse(xy=(x,y), width=5, height=2, angle=45, fc=c, lw=None, zorder=10)
    e13 = patches.Ellipse(xy=(x,y), width=5, height=2, angle=90, fc=c, lw=None, zorder=10)
    e14 = patches.Ellipse(xy=(x,y), width=5, height=2, angle=135, fc=c, lw=None, zorder=10)
    c1 = patches.Circle(xy=(x,y), radius=1, fc='y', lw=None, zorder=10)
    return [e11,e12,e13,e14,c1]

# def add_flowers(locs,offset,ax,n):
#     if locs==None:
#         for l in range(n):
#             #c = np.random.choice(['b','y','r','w','purple'])
#             c = (np.random.rand(),np.random.rand(),np.random.rand())
#             x = np.random.rand()*25
#             y = np.random.rand()*50
#             flower = make_flower(x+offset,y,c)
#             for x in flower: 
#                 ax.add_patch(x)
#     else:
#         for l in locs:
#             #c = np.random.choice(['b','y','r','w','purple'])
#             c = (np.random.rand(),np.random.rand(),np.random.rand())
#             flower = make_flower(l[0]+offset,l[1],c)
#             for x in flower: 
#                 ax.add_patch(x)

def make_flowers(locs,offset):
  flowers = []
  for l in locs:
    c = (np.random.rand(),np.random.rand(),np.random.rand())
    flowers.append(make_flower(l[0]+offset,l[1],c))
  return flowers

def make_saproling(x,y):
    s1 = patches.Rectangle(xy=(x-.5,y), width=1, height=3, fc='g', lw=None, zorder=9)
    s2 = patches.Ellipse(xy=(x-1.5,y+3.5), width=3, height=1, angle=150, fc='g', lw=None, zorder=9)
    s3 = patches.Ellipse(xy=(x+1.5,y+3.5), width=3, height=1, angle=30, fc='g', lw=None, zorder=9)
    return [s1,s2,s3]
    
def make_tree(x,y):
    t0 = patches.Rectangle(xy=(x,y), width=2, height=8, fc='brown', lw=None, zorder=10)
    t1 = patches.Ellipse(xy=(x+1,y+10), width=10, height=4, fc='g', lw=None, zorder=10)
    t2 = patches.Ellipse(xy=(x+1,y+10), width=10, height=4, angle=30, fc='g', lw=None, zorder=10)
    t3 = patches.Ellipse(xy=(x+1,y+10), width=10, height=4, angle=-30, fc='g', lw=None, zorder=10)
    t4 = patches.Ellipse(xy=(x+1,y+10), width=10, height=4, angle=60, fc='g', lw=None, zorder=10)
    t5 = patches.Ellipse(xy=(x+1,y+10), width=10, height=4, angle=-60, fc='g', lw=None, zorder=10)
    return [t0,t1,t2,t3,t4,t5]

def make_rock(x,y):
    r1 = patches.Polygon(trianglify(x,y,5), fc='grey', lw=None, zorder=100)
    return [r1]

def make_path(ax):
  x = np.linspace(0,50,100)
  y = -np.sin(x/8)*2+25
  ax.plot(y,x,c='#b3b3b3',linewidth=30)
  #ax.plot(y,x,c='#b3b3b3',linewidth=30,linestyle=':',dashes=(1, 0.5))

def make_background(c1,c2):
    colours = {'dead': '#4c5c4c', 'dormant': '#b9edb9', 'healthy': '#74f274','thriving':'#50ba50'}
    b1 = patches.Rectangle(xy=(0,0), width=25, height=50, fc=colours[c1], lw=None, zorder=0)
    b2 = patches.Rectangle(xy=(25,0), width=25, height=50, fc=colours[c2], lw=None, zorder=0)
    return [b1,b2]
    
fig = plt.figure(figsize=(6,6),dpi=100)
ax = fig.add_subplot()

ax.axes.xaxis.set_ticks([])
ax.axes.yaxis.set_ticks([])

ax.set_xlim([0, 50])
ax.set_ylim([0, 50])

lw_text = ax.text(0.05, 1.05,'left wing', ha="left",va="center", color='black', fontsize="medium", transform=ax.transAxes)
rw_text = ax.text(0.95, 1.05,'right wing', ha="right",va="center", color='black', fontsize="medium", transform=ax.transAxes)
        
l_count = ax.text(0.25, 0.05,f'read: {left_count}', ha="center",va="center", color='grey', fontsize="medium", transform=ax.transAxes,zorder=2000)
r_count = ax.text(0.75, 0.05,f'read: {right_count}', ha="center",va="center", color='grey', fontsize="medium", transform=ax.transAxes,zorder=2000)

l_target = ax.text(0.25, 0.95,f'target: {left_target}', ha="center",va="center", color='grey', fontsize="medium", transform=ax.transAxes)
r_target = ax.text(0.75, 0.95,f'target: {right_target}', ha="center",va="center", color='grey', fontsize="medium", transform=ax.transAxes)

l_title = ax.text(0.25, -0.05,cc(left_count).title(), ha="center",va="center", color='black', fontsize="large", transform=ax.transAxes)
r_title = ax.text(0.75, -0.05,cc(right_count).title(), ha="center",va="center", color='black', fontsize="large", transform=ax.transAxes)

l_description = ax.text(0.25, -0.15,description[cc(left_count)], wrap=True, ha="center",va="center", fontsize='x-small',color='black', transform=ax.transAxes)
r_description = ax.text(0.75, -0.15,description[cc(right_count)], wrap=True, ha="center",va="center", fontsize='x-small',color='black', transform=ax.transAxes)


background = make_background(cc(left_count),cc(right_count))
for x in background: 
  ax.add_patch(x)

path = make_path(ax)

def draw_side(s_count,ax,offset):
  #count_converter
  artists=[]
  ratio = s_count/(garden_target/2)
  if ratio <= 0.1:
    artists.append(make_rock(5+offset,10))
  elif ratio <= 0.2: 
    artists.append(make_rock(5+offset,10))
    artists.append(make_rock(15+offset,30))
  elif ratio <= 0.3: 
    artists.append(make_saproling(5+offset,10))
    artists.append(make_saproling(15+offset,30))
    artists.append(make_saproling(5+offset,40))
  elif ratio <= 0.4: 
    artists.append(make_saproling(5+offset,10))
    artists.append(make_saproling(15+offset,30))
    artists.append(make_saproling(5+offset,40))
    artists.append(make_saproling(10+offset,20))
  elif ratio <= 0.5: 
    artists.append(make_saproling(5+offset,10))
    artists.append(make_saproling(15+offset,30))
    artists.append(make_saproling(5+offset,40))
    artists.append(make_saproling(10+offset,20))
    artists.append(make_saproling(17+offset,5))
  elif ratio <= 0.6:
    locations = [[6,10],[16,17],[8,25],[18,33],[10,40]]
    for f in make_flowers(locations,offset): 
      artists.append(f)
  elif ratio <= 0.7:
    locations = [[18,5],[6,10],[16,17],[8,25],[18,33],[10,40]]
    for f in make_flowers(locations,offset): 
      artists.append(f)
  elif ratio <= 0.8:
    locations = [[20,45],[18,5],[6,10],[16,17],[8,25],[18,33],[10,40]]
    for f in make_flowers(locations,offset): 
      artists.append(f)
  elif ratio <= 0.9:
    locations = [[5,47],[20,45],[18,5],[6,10],[16,17],[8,25],[18,33],[10,40]]
    for f in make_flowers(locations,offset): 
      artists.append(f)
  elif ratio < 1:
    locations = [[11,31],[5,47],[20,45],[18,5],[6,10],[16,17],[7,22],[18,33],[10,40]]
    for f in make_flowers(locations,offset): 
      artists.append(f)
  elif ratio >= 1: 
    locations = [[11,31],[5,47],[20,45],[18,5],[6,10],[16,17],[7,22],[18,33],[10,40]]
    for f in make_flowers(locations,offset): 
      artists.append(f)
    artists.append(make_tree(5+offset,13))
    artists.append(make_tree(18+offset,22))
  return artists

lartists = draw_side(left_count,ax,0)
rartists = draw_side(right_count,ax,25)

for x in lartists: 
  for y in x:
    ax.add_patch(y)
for x in rartists: 
  for y in x:
    ax.add_patch(y)

plt.show()
```
